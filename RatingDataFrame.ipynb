```python
def urls2(urls,urls1):
    urls2 = list()
    i = 0
    for i in range(len(urls1)):
        x =urls[i]
        f = x.find(' ')
        if f > 0:
            urls2.append(urls1[i])
        else:
            urls2.append(urls[i])
    return urls2
```

```python
urls2 = urls2(urls,urls1)
```

```python
def urls3(urls2):
    urls3 = list()
    ul = "https://tabelog.com/rstLst/?vs=1&sa=&sk=" 
    ul2 ="&lid=hd_search1&vac_net=&svd=20171108&svt=1900&svps=2&hfc=1&sw=" 
    for i in range(len(urls2)):
        b = ul + urls2[i] + ul2 + urls2[i]
        urls3.append(b)
    return urls3
```

```python
urls3 = urls3(urls2)
```

```python
def bb(us):
    i = 0
    bb = list()
    for i in range(len(us)):
        session = requests.Session()
        response = session.get(us[i])
        root = lxml.html.fromstring(response.content)
        root.make_links_absolute(response.url)  
        for a in root.cssselect('a'):
            aa= a.get("href")
            d = list()
            if 'dtlrvwlst' in str(aa):
                bb.append(aa)
                d.append(aa)
                if len(d) > 0:
                    break            
    return bb
```

```python
bb = bb(urls3)
```

```python
def rate(bb):
    i = 0
    rate = list()
    for i in range(len(bb)):
        c = str(bb[i])
        cs = c.replace('dtlrvwlst', 'dtlratings')
        session = requests.Session()
        response = session.get(cs)
        root = lxml.html.fromstring(response.content)
        for x in root.cssselect('strong[class = "score"]'):
            rate.append(x.text)
    return rate
```

```python
rate = rate(bb)
```

```python
def seg(f1,f2,m):
    import pandas as pd
    from pandas import DataFrame as df
    import numpy as np
    i = 0
    f1g = np.array([])
    f2g = np.array([])
    mg = np.array([])
    for i in range(len(f1)):
        f = str(f1[i])
        if '女性' in f:
            if '2' or '3' in f:
                f1g = np.append(f1g, f)
                f1g = np.append(f1g, f2[i])
                f1g = np.append(f1g, m[i])
            else:
                f2g = np.append(f2g, f)
                f2g = np.append(f2g, f2[i])
                f2g = np.append(f2g, m[i])
        else:
            mg = np.append(mg, f)
            mg = np.append(mg, f2[i])
            mg = np.append(mg, m[i])
            
    value_df1 = pd.DataFrame(np.reshape(f1g,(int(len(f1g)/3),3)))
    value_df2 = pd.DataFrame(np.reshape(f2g,(int(len(f2g)/3),3)))
    value_dm = pd.DataFrame(np.reshape(mg,(int(len(mg)/3),3)))
    
    value_df1.columns = ['profile', 'price', 'rate']
    value_df2.columns = ['profile', 'price', 'rate']
    value_dm.columns = ['profile', 'price', 'rate']
    dff1 = value_df1
    dff2 = value_df2
    dfm = value_dm
    r1r = dff1['rate']
    r2r = dff2['rate']
    r3r = dfm['rate']
    r1 = np.array([])
    r4 = np.array([])
    r3 = np.array([])
    i =0
    for i in range(len(r1r)):
        r2 = r1r[i]
        if '-' in str(r2):
            pass
        else:
            r = float(r2)
            r1 = np.append(r1,r)
    i =0
    for i in range(len(r2r)):
        r2 = r2r[i]
        if '-' in str(r2):
            pass
        else:
            r = float(r2)
            r4 = np.append(r4,r)
    i =0
    for i in range(len(r3r)):
        r2 = r3r[i]
        if '-' in str(r2):
            pass
        else:
            r = float(r2)
            r3 = np.append(r3,r)

    ## 金額を抜き出す
    p1 = dff1['price']
    p2 = dff2['price']
    p3 = dfm['price']
    pf1 = np.array([])
    pf2 = np.array([])
    pm = np.array([])
    i = 0
    for i in range(len(p1)):
        p = str(p1[i])
        if len(p) == 13:
            d = int((int(p[1]+p[3:6])+int(p[8]+p[10:])+1)/2)
            pf1 = np.append(pf1,d)
        else:
            pass
    i = 0
    for i in range(len(p2)):
        p = str(p2[i])
        if len(p) == 13:
            d = (int(p[1]+p[3:6])+int(p[8]+p[10:])+1)/2
            pf2 = np.append(pf2,d)
        else:
            pass
    i = 0
    for i in range(len(p3)):
        p = str(p3[i])
        if len(p) == 13:
            d = (int(p[1]+p[3:6])+int(p[8]+p[10:])+1)/2
            pm = np.append(pm,d)
        else:
            pass 
        
    ## 比率を抜き出す
    i = 0
    c1 = list(dff1['profile'])
    c2 = list(dff2['profile'])
    c3 = list(dfm['profile'])
    c4 = list()
    c5 = list()
    for i in range(len(c3)):
        if c3[i] == 'None':
            pass 
        else:
            c5.append(c3[i])
    i = 0
    for i in range(len(c2)):
        c = c2[i]
        if '0' in c:
            c4.append(c)
        else:
            pass
    t1 = len(c1)/(len(c1)+len(c4)+len(c5))  
    t2 = len(c4)/(len(c1)+len(c4)+len(c5))  
    t3 = len(c5)/(len(c1)+len(c4)+len(c5))  
    np = list([t1,t3,pf1.mean(),pm.mean(),r1.mean(),r3.mean()])
    return np
```

```python
def usa1(cs):
    from pandas import DataFrame as df
    import numpy as np
    goal = np.array([])
    i = 0
    for i in range(len(cs)):
        nrate = list()
        pro = list()
        ra = list()
        nn = list()
        ni = list()
        session = requests.Session()
        response = session.get(cs[i])
        root = lxml.html.fromstring(response.content)
        for x in root.cssselect('span[class = "rvw-item__rvwr-profile"]'):
            pro.append(x.text)
        for x in root.cssselect('span'):
            nn.append(x.text)
        i = 0
        for i in range(len(nn)):
            n = str(nn[i])
            if "夜の点数：" in n:
                ni.append(n)
            if "昼の点数：" in n:
                ni.append(n)
            else:
                pass
        price = list()
        for x in root.cssselect('strong[class = "c-rating__val rvw-item__usedprice-price"]'):
            price.append(x.text)
        i = 0
        nr= list()
        while i < len(price) and i < len(ni):
            if "夜" in ni[i]:
                nr.append(price[i])
            else:
                pass
            i = i + 1
        trate= list()
        tr = list()
        for x in root.cssselect('b[class = "c-rating__val c-rating__val--strong"]'):
            trate.append(x.text)
        i = 0
        while i < len(trate) and i < len(ni):
            if "夜" in ni[i]:
                tr.append(trate[i])
            else:
                pass
            i = i + 1
        f1 = np.array(pro)
        f2 = np.array(nr)
        m = np.array(tr)     
        s = seg(f1,f2,m)
        goal = np.append(goal,s)
    return goal
```
